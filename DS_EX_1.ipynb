```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
a=pd.read_csv('Titanic-Dataset.csv')
```

```python
df=pd.DataFrame(a)
print(df)
```

```python
print(df.head(5))
```

```python
df.info()
```

```python
df['age_forward_fill'] = df['Age'].fillna(method='ffill')
df['age_backward_fill'] = df['Age'].fillna(method='bfill')
df
```

```python
df['Cabin'] = df['Cabin'].fillna("unknown", limit=5)
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['sex_encoded'] = df['Sex'].map({'male': 0, 'female': 1}).fillna('NaN')
print(df)
```

```python
scaler = StandardScaler()
df['Fare'] = scaler.fit_transform(df[['Fare']])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_features = ['Pclass', 'Sex', 'Age', 'SibSp']
sns.pairplot(df[selected_features])
plt.show()
```

```python
correlation_matrix = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
